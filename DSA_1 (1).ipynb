```python
def intostr(i):
  digit="0123456789"
  if i==0:
    return "0"
  result=" "
  while i>0:
    result=digit[i%10]+result
    i=i//10
  return result
```

```python
def factorial(n):
  if n==1:
    return 1
  else:
    return n*factorial(n-1)
```

```python
def fib(n):
  if n==1 or n==0:
    return 1
  else:
    return fib(n-1)+fib(n-2)
```
